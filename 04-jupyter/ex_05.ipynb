```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Data for plotting
t = np.arange(2.0, 10.0, 0.1)
s = 1 + np.sin(2 * np.pi * t)
```

```python
fig, ax = plt.subplots()
ax.plot(t, s)
```

```python
ax.set(xlabel='time (s)', ylabel='voltage (mV)', title='About as simple as it gets, folks')
ax.grid()
```

```python
plt.show()
```
